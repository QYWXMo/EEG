```python
import scipy.io as scio
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew, kurtosis
from scipy.signal import welch
from scipy.signal import stft
import matplotlib.pyplot as plt


# 
data_array = data['data']

# 
print(":", data_array.shape)

#
sampling_rate = 1000  # 1000Hz
num_samples = data_array.shape[1]  # 
duration_in_seconds = num_samples / sampling_rate
print(":", duration_in_seconds)

import numpy as np
from scipy.stats import skew, kurtosis

# 
def extract_time_domain_features(signal):
    features = {}
    features['mean'] = np.mean(signal)
    features['std'] = np.std(signal)
    features['max'] = np.max(signal)
    features['min'] = np.min(signal)
    features['peak_to_peak'] = np.ptp(signal)  # Peak-to-Peak
    features['variance'] = np.var(signal)
    features['skewness'] = skew(signal)
    features['kurtosis'] = kurtosis(signal)
    return features

# 16
def aggregate_features_across_channels(data_array):
    aggregated_features = {
        'mean': [],
        'std': [],
        'max': [],
        'min': [],
        'peak_to_peak': [],
        'variance': [],
        'skewness': [],
        'kurtosis': []
    }

    # 16
    for i in range(data_array.shape[0]):
        channel_signal = data_array[i, :]
        channel_features = extract_time_domain_features(channel_signal)

        # 
        for key in aggregated_features:
            aggregated_features[key].append(channel_features[key])

    # 16
    for key in aggregated_features:
        aggregated_features[key] = {
            'mean_across_channels': np.mean(aggregated_features[key]),
            'std_across_channels': np.std(aggregated_features[key])
        }

    return aggregated_features

# 16
time_domain_features_across_channels = aggregate_features_across_channels(data_array)
print(time_domain_features_across_channels)


import numpy as np
import pandas as pd
from scipy.signal import welch

# 
def extract_frequency_domain_features(signal, fs):
    # 
    freqs, psd = welch(signal, fs=fs, nperseg=1024)
    
    # 
    delta_power = np.sum(psd[(freqs >= 0.5) & (freqs < 4)])    # Delta (0.5-4 Hz)
    theta_power = np.sum(psd[(freqs >= 4) & (freqs < 8)])      # Theta (4-8 Hz)
    alpha_power = np.sum(psd[(freqs >= 8) & (freqs < 13)])     # Alpha (8-13 Hz)
    beta_power = np.sum(psd[(freqs >= 13) & (freqs < 30)])     # Beta (13-30 Hz)
    gamma_power = np.sum(psd[(freqs >= 30) & (freqs < 100)])   # Gamma (30-100 Hz)
    
    frequency_features = {
        'delta_power': delta_power,
        'theta_power': theta_power,
        'alpha_power': alpha_power,
        'beta_power': beta_power,
        'gamma_power': gamma_power
    }
    return frequency_features

#  (16, 604700) fs  1000 Hz
fs = 1000  # 
n_channels = data_array.shape[0]

# 
frequency_domain_features = []
for i in range(n_channels):
    features = extract_frequency_domain_features(data_array[i, :], fs)
    frequency_domain_features.append(features)

#  DataFrame
df_frequency_features = pd.DataFrame(frequency_domain_features)

# 
mean_features = {
    'delta_power_mean_across_channels': np.mean(df_frequency_features['delta_power']),
    'theta_power_mean_across_channels': np.mean(df_frequency_features['theta_power']),
    'alpha_power_mean_across_channels': np.mean(df_frequency_features['alpha_power']),
    'beta_power_mean_across_channels': np.mean(df_frequency_features['beta_power']),
    'gamma_power_mean_across_channels': np.mean(df_frequency_features['gamma_power'])
}

std_features = {
    'delta_power_std_across_channels': np.std(df_frequency_features['delta_power']),
    'theta_power_std_across_channels': np.std(df_frequency_features['theta_power']),
    'alpha_power_std_across_channels': np.std(df_frequency_features['alpha_power']),
    'beta_power_std_across_channels': np.std(df_frequency_features['beta_power']),
    'gamma_power_std_across_channels': np.std(df_frequency_features['gamma_power'])
}

# 
for i in range(n_channels):
    print(f"\n {i+1} :")
    print(df_frequency_features.iloc[i])

# 
print("\n:")
for band, mean_value in mean_features.items():
    print(f"{band}: {mean_value}")

print("\n:")
for band, std_value in std_features.items():
    print(f"{band}: {std_value}")



# 
channel_1 = data_array[0, :]

# 
plt.figure(figsize=(12, 6))
plt.plot(channel_1[:10000])  # 10000
plt.title('Channel 1 - Time Domain Signal')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.show()


```

```python
import pywt
# 
#  (STFT)
channel_1 = data_array[0, :]
fs = 1000
f, t, Zxx = stft(channel_1, fs, nperseg=1000)

# STFT
plt.pcolormesh(t, f, np.abs(Zxx), shading='gouraud')
plt.title('STFT Magnitude')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()

#  (CWT)
coefficients, frequencies = pywt.cwt(channel_1, scales=np.arange(1, 128), wavelet='cmor', sampling_period=1/fs)

# 
plt.figure(figsize=(12, 6))
plt.imshow(np.abs(coefficients), extent=[0, len(channel_1)/fs, 1, 128], cmap='PRGn', aspect='auto',
           vmax=abs(coefficients).max(), vmin=-abs(coefficients).max())
plt.title('Wavelet Transform (CWT)')
plt.ylabel('Scale')
plt.xlabel('Time [sec]')
plt.show()
```
